```python
from sklearn.datasets import load_iris
iris = load_iris() # 아이리스 붓꽃 데이터

import pandas as pd
# 데이터 셋을 DataFrame으로 변환(확인용)
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
#print(df)

features = iris.data # 데이터
feature_names = iris.feature_names # 열 제목

import numpy as np
sepal_length=np.array(features[:,0])
sepal_width=np.array(features[:,1])
petal_length=np.array(features[:,2])
petal_width=np.array(features[:,3])

sepal_length_scaled=(sepal_length - np.mean(sepal_length))/np.std(sepal_length)
sepal_width_scaled=(sepal_width - np.mean(sepal_width))/np.std(sepal_width)
petal_length_scaled=(petal_length - np.mean(petal_length))/np.std(petal_length)
petal_width_scaled=(petal_width - np.mean(petal_width))/np.std(petal_width)

import matplotlib.pyplot as plt

plt.figure(figsize=(6.5,10))

plt.subplot(2, 1, 1)
plt.scatter(sepal_length_scaled,sepal_width_scaled,color='red')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])

plt.subplot(2, 1, 2)
plt.scatter(petal_length_scaled,petal_width_scaled,color='blue')
plt.xlabel(feature_names[2])
plt.ylabel(feature_names[3])
plt.show()
```

```python
from sklearn import datasets,preprocessing
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

wine = datasets.load_wine(as_frame=True)

'''
features = wine.data # 데이터
feature_names = wine.feature_names # 열 제목
print(features)
print(feature_names)
'''

X_scaled = preprocessing.scale(wine.data)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, wine.target, test_size=0.25)
# test_size 매개변수 : 훈련 데이터를 테스트 세트로 나눌 비율 => 기본값 = 0.25

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
# n_neighbors 매개변수 : 주변 가까운 이웃의 개수 지정

knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```
